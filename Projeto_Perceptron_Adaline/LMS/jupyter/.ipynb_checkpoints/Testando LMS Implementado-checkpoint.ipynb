```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
from random import randint

sys.path.append("/home/dimi/Programming/MachineLearningUFABC/Projeto_Perceptron_Adaline/LMS/python/")
from LMSClass import LMS

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
params = {'figure.figsize': [10, 5], 
          'figure.titlesize': 16,
          'axes.labelsize': 12,
          'axes.titlesize':14, 
          'font.size': 12,
          'legend.fontsize': 12, 
          'xtick.labelsize': 10, 
          'ytick.labelsize': 10
         }
plt.rcParams.update(params)
```

```python
for i in range(10):

    # COEFICIENTES DA RETA ALEATORIOS
    coeficienteAngular = randint(-10,10)
    coeficienteLinear  = randint(-10,10)

    # CRIANDO O DATASET COM QUANTIDADE E RANGE DE PONTOS ALEATORIOS
    x = np.linspace(-randint(500,1000), randint(500,1000), randint(50,100))
    y = []
    for entradaAtual in x:
        y.append(coeficienteAngular*(entradaAtual + randint(-250,250)) + coeficienteLinear)
    y = np.array(y)

    # USANDO A CLASSE IMPLEMENTADA
    objLMS = LMS(x, y)
    print("MSE:", objLMS.MSE)
    print("RMSE:", objLMS.RMSE)

    # PLOTANDO
    eixoXRetaLMS, eixoYRetaLMS = objLMS.obterEixosParaPlotarRetaLMS(x)
    plt.scatter(x, y, s=10)
    plt.plot(eixoXRetaLMS, eixoYRetaLMS, color="tab:brown")
    plt.grid(alpha=0.5)
    plt.xlim(-1000, 1000)
    plt.ylim(-10010, 10010)
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
from random import randint

sys.path.append("/home/dimi/Programming/MachineLearningUFABC/Projeto_Perceptron_Adaline/LMS/python/")
from LMSClass import LMS

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
params = {'figure.figsize': [10, 5], 
          'figure.titlesize': 16,
          'axes.labelsize': 12,
          'axes.titlesize':14, 
          'font.size': 12,
          'legend.fontsize': 12, 
          'xtick.labelsize': 10, 
          'ytick.labelsize': 10
         }
plt.rcParams.update(params)
```

```python
for i in range(10):

    # COEFICIENTES DA RETA ALEATORIOS
    coeficienteAngular = randint(-5,5)
    coeficienteLinear  = randint(-5,5)

    # CRIANDO O DATASET COM QUANTIDADE E RANGE DE PONTOS ALEATORIOS
    x = np.linspace(-randint(50,100), randint(50,100), randint(25,75))
    y = []
    for entradaAtual in x:
        y.append(coeficienteAngular*(entradaAtual + randint(-20,20)) + coeficienteLinear)
    y = np.array(y)

    # USANDO A CLASSE IMPLEMENTADA
    objLMS = LMS(x, y)
    print("Pesos Reais:", [coeficienteLinear, coeficienteAngular])
    print("Pesos calculados:", objLMS.pesos)
    print("MSE:", objLMS.MSE)
    print("RMSE:", objLMS.RMSE)

    # PLOTANDO
    eixoXRetaLMS, eixoYRetaLMS = objLMS.obterEixosParaPlotarRetaLMS(x)
    plt.scatter(x, y, s=10)
    plt.plot(eixoXRetaLMS, eixoYRetaLMS, color="tab:brown")
    plt.grid(alpha=0.5)
    plt.xlim(-105, 105)
    plt.ylim(-510, 510)
    plt.show()
```

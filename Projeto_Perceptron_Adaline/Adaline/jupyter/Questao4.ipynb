## Importando os pacotes necessários

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
import time
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import mean_squared_error
from random import randint

sys.path.append("/home/dimi/Programming/MachineLearningUFABC/Projeto_Perceptron_Adaline/Adaline/python/")
sys.path.append("/home/dimi/Programming/MachineLearningUFABC/Projeto_Perceptron_Adaline/LMS/python/")
from PolinomialAdalineClass import PolinomialAdaline
from PolinomialLMSClass import PolinomialLMS

pd.options.mode.chained_assignment = None

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
params = {'figure.figsize': [12, 3], 
          'figure.titlesize': 14,
          'axes.labelsize': 10,
          'axes.titlesize':12, 
          'font.size': 10,
          'legend.fontsize': 10, 
          'xtick.labelsize': 8, 
          'ytick.labelsize': 8
         }
plt.rcParams.update(params)
```

```python
def plotarDatasetsPolinomiais(x, y, objAdaline, objLMS):
    print("Pesos calculados pelo LMS:", objLMS.pesos)
    print("Pesos calculados pelo Adaline:", objAdaline.pesos)

    eixoXRetaAdaline, eixoYRetaAdaline = objAdaline.obterEixosParaPlotarRetaAdaline(x)
    eixoXRetaLMS, eixoYRetaLMS         = objLMS.obterEixosParaPlotarRetaLMS(x)

    fig, axs = plt.subplots(1, 2)

    axs[0].scatter(x, y, s=10, color="tab:grey", label="Dados")
    axs[0].plot(eixoXRetaAdaline, eixoYRetaAdaline, "--", label="Adaline")
    axs[0].plot(eixoXRetaLMS, eixoYRetaLMS, "--", label="LMS")
    axs[0].set_title("Retas encontradas pelos algoritmos")
    axs[0].legend()
    axs[0].grid(alpha=0.5)

    axs[1].plot(objAdaline.evolucaoMSE, label="Adaline")
    axs[1].grid(alpha=0.5)
    axs[1].set_title("Evolução dos MSE")
    axs[1].set_xlabel("Época")
    axs[1].set_ylabel("MSE")

    plt.suptitle("Grau: " + str(len(objAdaline.pesos)-1))
    plt.show()
    
```

## Criando datasets polinomiais

```python
for grau in range(2, 10):

    # GERANDO OS DADOS
    x = np.linspace(-10, 10, 50)
    y = 0
    pesosOriginais = []

    for i in range(0, grau+1):
        coeficienteAtual = randint(-10, 10)
        pesosOriginais.append(coeficienteAtual)
        y += coeficienteAtual*x**(i)

    # ADALINE
    taxaAprendizagem = 1*10**(-(2*grau+1))
    print("Pesos originais:", pesosOriginais)
    objAdaline = PolinomialAdaline(grau)
    objAdaline.treinar(x, y, taxaAprendizagem, qtdMaxEpocas=100)
    
    # LMS
    objLMS = PolinomialLMS(x, y, grau)    

    plotarDatasetsPolinomiais(x, y, objAdaline, objLMS)
```

## Importando o dataset da atividade

```python
dataframe = pd.read_csv("/home/dimi/Programming/MachineLearningUFABC/Projeto_Perceptron_Adaline/Datasets/dataset_regression.csv", sep=";", header=None, names=["y","x"])
x = dataframe["x"].to_numpy()
y = dataframe["y"].to_numpy()
```

### Solução com LMS

```python
arrayMSExGrau = []
rangeGraus = np.arange(1,13)

for grau in rangeGraus:
    
    objLMS = PolinomialLMS(x, y, grau) 
    eixoXRetaLMS, eixoYRetaLMS = objLMS.obterEixosParaPlotarRetaLMS(x)
    arrayMSExGrau.append(objLMS.MSE)

    plt.scatter(x, y, s=10, color="tab:grey", label="Dados")
    plt.plot(eixoXRetaLMS, eixoYRetaLMS, "--", label="LMS")
    plt.title("Grau LMS: " + str(grau))
    plt.legend()
    plt.grid(alpha=0.5)
    plt.show()
    
plt.plot(rangeGraus, arrayMSExGrau, color="tab:red")
plt.title("MSE em função do grau para o LMS polinomial")
plt.grid(alpha=0.5)
plt.show()
```

### Solução com Adaline

```python
grau = 10

objAdaline = PolinomialAdaline(grau)
objAdaline.treinar(x, y, taxaAprendizagem=1e-49, qtdMaxEpocas=1000000)

objLMS = PolinomialLMS(x, y, grau)    

plotarDatasetsPolinomiais(x, y, objAdaline, objLMS)
```

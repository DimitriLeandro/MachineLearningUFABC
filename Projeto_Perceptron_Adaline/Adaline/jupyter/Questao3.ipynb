## Sobre as Implementações

Nessa etapa, duas classes foram implementadas: **LMS** e **Adaline**. 

A primeira encontra os pesos de cada entrada, incluindo o bias, pela expressão determinística $w = (\chi^T \cdot \chi)^{-1} \cdot \chi^T \cdot y^T$. 

Já a segunda classe conta com o algoritmo do neurônio, mas sem função de ativação. Ela busca minímizar o Erro Quadrático Médio (MSE) das predições ao longo de várias épocas de treinamento. Os critérios de parada implementados foram: quantidade máxima de épocas de treinamento e percentual desse total sem melhorias no MSE.

## Importando os pacotes necessários

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
import time
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import mean_squared_error
from random import randint

sys.path.append("/home/dimi/Programming/MachineLearningUFABC/Projeto_Perceptron_Adaline/LMS/python/")
sys.path.append("/home/dimi/Programming/MachineLearningUFABC/Projeto_Perceptron_Adaline/Adaline/python/")
from LMSClass import LMS
from AdalineClass import Adaline

pd.options.mode.chained_assignment = None

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
params = {'figure.figsize': [10, 5], 
          'figure.titlesize': 16,
          'axes.labelsize': 12,
          'axes.titlesize':14, 
          'font.size': 12,
          'legend.fontsize': 12, 
          'xtick.labelsize': 10, 
          'ytick.labelsize': 10
         }
plt.rcParams.update(params)
```

## Importando o dataset

```python
dataframe = pd.read_csv("/home/dimi/Programming/MachineLearningUFABC/Projeto_Perceptron_Adaline/Datasets/dataset_regression.csv", sep=";", header=None, names=["y","x"])
x = dataframe["x"].to_numpy()
y = dataframe["y"].to_numpy()
plt.scatter(x, y, s=10)
plt.grid(alpha=0.5)
plt.show()
```

## Solução com LMS

```python
inicioLMS = time.time()
objLMS    = LMS(x, y)    
fimLMS    = time.time()

print("Tempo de Processamento do LMS (s):", fimLMS - inicioLMS)
print("Pesos calculados pelo LMS:", objLMS.pesos)
print("MSE do LMS:", objLMS.MSE)
print("RMSE do LMS:", objLMS.RMSE)
```

## Solução com Adaline

Nesta seção, diferentes taxas de aprendizagem serão testadas e comparadas. Testes prévios mostraram que usando qualquer taxa maior ou igual a 1e-4, os pesos começam a crescer demais e ocorre um problema de overflow. Por isso, os testes incluirão apenas taxas menores ou iguais a 1e-5.

### Taxa de 1e-5

```python
inicioAdaline1e5 = time.time()
objAdaline1e5    = Adaline()
objAdaline1e5.treinar(x, y, taxaAprendizagem=1e-5)   
fimAdaline1e5    = time.time()

print("Tempo de Processamento do Adaline (s):", fimAdaline1e5 - inicioAdaline1e5)
print("Pesos calculados pelo Adaline:", objAdaline1e5.pesos)
print("MSE do Adaline:", min(objAdaline1e5.evolucaoMSE))
print("RMSE do Adaline:", min(objAdaline1e5.evolucaoMSE)**(1/2))
```

### Taxa de 1e-6

```python
inicioAdaline1e6 = time.time()
objAdaline1e6    = Adaline()
objAdaline1e6.treinar(x, y, taxaAprendizagem=1e-6)   
fimAdaline1e6    = time.time()

print("Tempo de Processamento do Adaline (s):", fimAdaline1e6 - inicioAdaline1e6)
print("Pesos calculados pelo Adaline:", objAdaline1e6.pesos)
print("MSE do Adaline:", min(objAdaline1e6.evolucaoMSE))
print("RMSE do Adaline:", min(objAdaline1e6.evolucaoMSE)**(1/2))
```

### Taxa de 1e-7

```python
inicioAdaline1e7 = time.time()
objAdaline1e7    = Adaline()
objAdaline1e7.treinar(x, y, taxaAprendizagem=1e-7)   
fimAdaline1e7    = time.time()

print("Tempo de Processamento do Adaline (s):", fimAdaline1e7 - inicioAdaline1e7)
print("Pesos calculados pelo Adaline:", objAdaline1e7.pesos)
print("MSE do Adaline:", min(objAdaline1e7.evolucaoMSE))
print("RMSE do Adaline:", min(objAdaline1e7.evolucaoMSE)**(1/2))
```

## Comparação

```python
eixoXRetaLMS, eixoYRetaLMS               = objLMS.obterEixosParaPlotarRetaLMS(x)
eixoXRetaAdaline1e5, eixoYRetaAdaline1e5 = objAdaline1e5.obterEixosParaPlotarRetaAdaline(x)
eixoXRetaAdaline1e6, eixoYRetaAdaline1e6 = objAdaline1e6.obterEixosParaPlotarRetaAdaline(x)
eixoXRetaAdaline1e7, eixoYRetaAdaline1e7 = objAdaline1e7.obterEixosParaPlotarRetaAdaline(x)

plt.scatter(x, y, s=10, color="tab:grey", label="Dados")
plt.plot(eixoXRetaLMS, eixoYRetaLMS, "--", label="LMS")
plt.plot(eixoXRetaAdaline1e5, eixoYRetaAdaline1e5, "--", label="Adaline 1e-5")
plt.plot(eixoXRetaAdaline1e6, eixoYRetaAdaline1e6, "--", label="Adaline 1e-6")
plt.plot(eixoXRetaAdaline1e7, eixoYRetaAdaline1e7, "--", label="Adaline 1e-7")
plt.title("Retas encontradas pelos algoritmos")
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

```python
plt.plot([0,len(objAdaline1e5.evolucaoMSE)], [objLMS.MSE, objLMS.MSE], "--", label="LMS")
plt.plot(objAdaline1e5.evolucaoMSE, label="Adaline 1e-5")
plt.plot(objAdaline1e6.evolucaoMSE, label="Adaline 1e-6")
plt.plot(objAdaline1e7.evolucaoMSE, label="Adaline 1e-7")
plt.legend()
plt.grid(alpha=0.5)
plt.title("Evolução dos MSE")
plt.xlabel("Época")
plt.ylabel("MSE")
plt.show()
```

```python
plt.bar("LMS" ,fimLMS-inicioLMS)
plt.bar("Adaline 1e-5" ,fimAdaline1e5-inicioAdaline1e5)
plt.bar("Adaline 1e-6" ,fimAdaline1e6-inicioAdaline1e6)
plt.bar("Adaline 1e-7" ,fimAdaline1e7-inicioAdaline1e7)
plt.title("Tempo de processamento dos algoritmos")
plt.ylabel("Tempo (s)")
plt.yscale("log")
plt.grid(alpha=0.5, axis="y")
plt.show()
```

## 50% dos dados para treinamento

### Treinamento

```python
xTrain, xTest, yTrain, yTest = tts(x, y, test_size=0.5, shuffle=True)

objLMS        = LMS(xTrain, yTrain)
objAdaline1e5 = Adaline()
objAdaline1e5.treinar(xTrain, yTrain, taxaAprendizagem=1e-5)
objAdaline1e6 = Adaline()
objAdaline1e6.treinar(xTrain, yTrain, taxaAprendizagem=1e-6)
objAdaline1e7 = Adaline()
objAdaline1e7.treinar(xTrain, yTrain, taxaAprendizagem=1e-7)
```

### Teste

```python
MSETesteLMS        = mean_squared_error(yTest, objLMS.predizer(xTest))
MSETesteAdaline1e5 = mean_squared_error(yTest, objAdaline1e5.predizer(xTest))
MSETesteAdaline1e6 = mean_squared_error(yTest, objAdaline1e6.predizer(xTest))
MSETesteAdaline1e7 = mean_squared_error(yTest, objAdaline1e7.predizer(xTest))
```

### Comparação

```python
eixoXRetaLMS, eixoYRetaLMS               = objLMS.obterEixosParaPlotarRetaLMS(x)
eixoXRetaAdaline1e5, eixoYRetaAdaline1e5 = objAdaline1e5.obterEixosParaPlotarRetaAdaline(x)
eixoXRetaAdaline1e6, eixoYRetaAdaline1e6 = objAdaline1e6.obterEixosParaPlotarRetaAdaline(x)
eixoXRetaAdaline1e7, eixoYRetaAdaline1e7 = objAdaline1e7.obterEixosParaPlotarRetaAdaline(x)

plt.scatter(xTrain, yTrain, s=10, alpha=0.75, color="tab:grey", label="Dados de treinamento")
plt.scatter(xTest, yTest, s=10, alpha=0.75, color="black", label="Dados de teste")
plt.plot(eixoXRetaLMS, eixoYRetaLMS, "--", label="LMS")
plt.plot(eixoXRetaAdaline1e5, eixoYRetaAdaline1e5, "--", label="Adaline 1e-5")
plt.plot(eixoXRetaAdaline1e6, eixoYRetaAdaline1e6, "--", label="Adaline 1e-6")
plt.plot(eixoXRetaAdaline1e7, eixoYRetaAdaline1e7, "--", label="Adaline 1e-7")
plt.title("Retas encontradas pelos algoritmos (50% para treinamento)")
plt.legend()
plt.grid(alpha=0.5)
plt.show()
```

```python
plt.bar(1, objLMS.MSE, color="tab:blue", label="MSE de treinamento")
plt.bar(2, MSETesteLMS, color="tab:orange", label="MSE de teste")

plt.bar(4, min(objAdaline1e5.evolucaoMSE), color="tab:blue")
plt.bar(5, MSETesteAdaline1e5, color="tab:orange")

plt.bar(7, min(objAdaline1e6.evolucaoMSE), color="tab:blue")
plt.bar(8, MSETesteAdaline1e6, color="tab:orange")

plt.bar(10, min(objAdaline1e7.evolucaoMSE), color="tab:blue")
plt.bar(11, MSETesteAdaline1e7, color="tab:orange")

plt.title("MSE de treinamento e teste dos algoritmos (50% para treinamento)")
plt.legend()
plt.xticks([1.5, 4.5, 7.5, 10.5], ["LMS", "Adaline 1e-5", "Adaline 1e-6", "Adaline 1e-7"])
plt.ylabel("MSE")
plt.grid(alpha=0.5, axis="y")
plt.show()
```

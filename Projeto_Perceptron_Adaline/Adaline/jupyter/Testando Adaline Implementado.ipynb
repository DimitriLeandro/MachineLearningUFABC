```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
from random import randint

sys.path.append("/home/dimi/Programming/MachineLearningUFABC/Projeto_Perceptron_Adaline/LMS/python/")
sys.path.append("/home/dimi/Programming/MachineLearningUFABC/Projeto_Perceptron_Adaline/Adaline/python/")
from LMSClass import LMS
from AdalineClass import Adaline

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
params = {'figure.figsize': [10, 5], 
          'figure.titlesize': 16,
          'axes.labelsize': 12,
          'axes.titlesize':14, 
          'font.size': 12,
          'legend.fontsize': 12, 
          'xtick.labelsize': 10, 
          'ytick.labelsize': 10
         }
plt.rcParams.update(params)
```

```python
for i in range(5):

    # COEFICIENTES DA RETA ALEATORIOS
    coeficienteAngular = randint(-5,5)
    coeficienteLinear  = randint(-5,5)
    print("Pesos Reais:", [coeficienteLinear, coeficienteAngular])

    # CRIANDO O DATASET COM QUANTIDADE E RANGE DE PONTOS ALEATORIOS
    x = np.linspace(-randint(50,100), randint(50,100), randint(25,75))
    y = []
    for entradaAtual in x:
        y.append(coeficienteAngular*(entradaAtual + randint(-20,20)) + coeficienteLinear)
    y = np.array(y)

    # USANDO A CLASSE IMPLEMENTADA DO LMS
    objLMS = LMS(x, y)    
    print("Pesos calculados pelo LMS:", objLMS.pesos)
    print("MSE do LMS:", objLMS.MSE)
    
    # USANDO A CLASSE IMPLEMENTADA DO ADALINE
    objAdaline = Adaline()
    objAdaline.treinar(x, y, taxaAprendizagem=1e-5)
    print("Pesos calculados pelo Adaline:", objAdaline.pesos)
    #print("MSE do LMS:", objLMS.MSE)

    # PLOTANDO
    eixoXRetaLMS, eixoYRetaLMS = objLMS.obterEixosParaPlotarRetaLMS(x)
    eixoXRetaAdaline, eixoYRetaAdaline = objAdaline.obterEixosParaPlotarRetaAdaline(x)
    plt.scatter(x, y, s=10, color="tab:grey")
    plt.plot(eixoXRetaLMS, eixoYRetaLMS, color="tab:red", label="LMS")
    plt.plot(eixoXRetaAdaline, eixoYRetaAdaline, "--", color="tab:blue", label="Adaline")
    plt.legend()
    plt.grid(alpha=0.5)
    plt.xlim(-105, 105)
    plt.ylim(-510, 510)
    plt.show()
```

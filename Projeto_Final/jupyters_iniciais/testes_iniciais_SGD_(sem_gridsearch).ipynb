```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score
```

```python
#DEFININDO ALGUNS PARÂMETROS DO GRÁFICO
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
params = {'figure.figsize': [11, 5.5], 
          'axes.labelsize': 18,
          'axes.titlesize':20, 
          'font.size': 18,
          'legend.fontsize': 18, 
          'xtick.labelsize': 14, 
          'ytick.labelsize': 14,
          'axes.axisbelow': True
         }

plt.rcParams.update(params)
```

### Abrindo o dataframe de teste

```python
dfTeste = pd.read_csv("/home/dimi/Programming/MachineLearningUFABC/Projeto_Final/dataset_teste.csv")
xTest   = dfTeste.iloc[:, 5:-1].to_numpy()
yTest   = dfTeste.iloc[:, -1].to_numpy()

print("Shape do xTeste:", xTest.shape)
print("Shape do yTeste:", yTest.shape)
```

### Setando parâmetros iniciais

```python
# VAMOS FAZER UM BOOTSTRAP COM 5 HOLDOUTS
qtdIteracoesCV = 10

# PERCENTUAL DE DADOS DO DATASET DE TESTE PRA USAR EM CADA ITERACAO
percentualTeste = 2/3
```

# Testando cada dataset de treinamento

### Treinamento original

```python
# ABRINDO O DATASET DE TREINAMENTO E SEPARANDO X E Y
arquivoCSV = "dataset_treinamento_original.csv"
dfTreinamento = pd.read_csv("/home/dimi/Programming/MachineLearningUFABC/Projeto_Final/"+arquivoCSV)
xTrain = dfTreinamento.iloc[:, 5:-1].to_numpy()
yTrain = dfTreinamento.iloc[:, -1].to_numpy()

# CRIANDO OS ARRAY DE ACURACIAS QUE VAI GUARDAR CADA ACURACIA DO CROSS VALIDATION
arrayAcuraciasOriginal = []

# ITERANDO O CROSS VALIDATION
for i in range(1, qtdIteracoesCV + 1):
    
    # NOS JA TEMOS XTRAIN E YTRAIN, PRECISO DAR UM SHUFFLE NOS DADOS DE TESTE
    xTest, yTest = shuffle(xTest, yTest)
    xTestAtual = xTest[:int(percentualTeste*len(xTest))]
    yTestAtual = yTest[:int(percentualTeste*len(yTest))]
    
    # CRIANDO, TREINANDO E PREDIZENDO O MODELO
    objSGD = SGDClassifier(n_jobs=-1).fit(xTrain, yTrain)
    yPred  = objSGD.predict(xTestAtual)
    arrayAcuraciasOriginal.append(accuracy_score(yTestAtual, yPred))
    
    # PRINTANDO O PROGRESSO
    print(str(100*(i/qtdIteracoesCV))+"%", end=" ")
    
print("\n" + str(arrayAcuraciasOriginal))
```

## Importando um datset não binário

```python
data   = load_wine().data
target = load_wine().target
xTrain, xTest, yTrain, yTest = tts(data, target, test_size=0.25, shuffle=True)
```

## Implementando

#### Função para criar um target binário

Um neurônio vai ser criado para cada classe do dataset. Cada neurônio vai ser treinado como "Verifique se é da classe atual ou se é de qualquer outra classe". Dado uma classe, essa função vai transformar o vetor target em 0s e 1s, sendo 1 pertencendo à classe desejada e 0 qualquer outra.

```python
def transformarTarget(target, classeDesejada):
    
    novoTarget = []
    
    for rotuloAtual in target:
        if rotuloAtual == classeDesejada:
            novoTarget.append(1)
        else:
            novoTarget.append(0)
            
    return np.array(novoTarget)
```

#### Função para criar e treinar um único neurônio dada uma classe

```python
def criarETreinarUnicoNeuronio(xTreinamento, yTreinamento, classeDesejada):
    
    # CRIANDO UM DATASET DO TIPO "CLASSE ATUAL (1) X QQR OUTRA CLASSE (0)"
    yBinarioClasseAtual = transformarTarget(yTreinamento, classeDesejada)
    
    # INSTANCIANDO E TREINANDO O NEURONIO ATUAL
    objNeuronio = Neuronio()
    objNeuronio.treinar(xTreinamento, yBinarioClasseAtual, armazenarEvolucao=True)
    
    return objNeuronio
```

```python
%%time

# VERIFICANDO QUAIS SAO AS CLASSES EXISTENTES NO DATASET
arrayClasses = np.unique(yTrain)

# CRIANDO UM NEURONIO BINARIO PARA CADA CLASSE DO DATASET
arrayNeuronios = []
for i, classeAtual in enumerate(arrayClasses):
    
    print("Iniciando treinamento do neurônio", i+1)
    
    objNeuronioAtual = criarETreinarUnicoNeuronio(xTrain, yTrain, classeAtual)

    # COLOCANDO NO ARRAY DE NEURONIOS
    arrayNeuronios.append(objNeuronioAtual)
    
print("Fim")
```

#### Função para treinar o algoritmo do perceptron multiclasse

```python
def treinar(xTreinamento, yTreinamento):
    
    # VERIFICANDO QUAIS SAO AS CLASSES EXISTENTES NO DATASET
    arrayClasses = np.unique(yTrain)

    # CRIANDO E TREINANDO UM NEURONIO PARA CADA CLASSE DO DATASET
    arrayNeuronios = Parallel(n_jobs=-1)(delayed(criarETreinarUnicoNeuronio)(xTreinamento, yTreinamento, classeAtual) for classeAtual in arrayClasses)
    
    return arrayNeuronios
```

```python
%%time

arrayNeuronios = treinar(xTrain, yTrain)
```

#### Função para devolver o index do neuronio com a maior entrada líquida para um único dado de entrada

```python
def maiorEntradaLiquida(dado, arrayNeuronios):
    
    # ADICIONANDO BIAS AO FINAL DO DADO
    dadoComBias = np.append(dado, 1)
    
    # VOU VERIFICAR A ENTRADA LIQUIDA DE CADA NEURONIO, QUEM TIVER A MAIOR GANHA
    arrayEntradasLiquidas = []
    for i, neuronioAtual in enumerate(arrayNeuronios):
        arrayEntradasLiquidas.append(neuronioAtual.calcularEntradaLiquida(dadoComBias))
        
    # RETORNANDO O INDEX DO NEURONIO COM A MAIOR ENTRADA LIQUIDA COMO A CLASSE PREDITA
    return arrayEntradasLiquidas.index(max(arrayEntradasLiquidas))
```

#### Função para predizer

```python
def predizer(xTeste):    
    arrayPredicoes = Parallel(n_jobs=-1)(delayed(maiorEntradaLiquida)(dadoAtual, arrayNeuronios) for dadoAtual in xTeste)
    return arrayPredicoes
```

```python
yPred = predizer(xTest)
accuracy_score(yTest, yPred)
```

# Implementação do Perceptron

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from random import randint
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split as tts
from sklearn.datasets import load_wine, load_breast_cancer
```

## Importando o dataset

```python
data   = load_breast_cancer().data
target = load_breast_cancer().target

xTrain, xTest, yTrain, yTest = tts(data, target, test_size=0.25, shuffle=True)
```

## Definindo funções para um único neurônio

As funções abaixo são para um único neurônio, isto é, um classificador binário, uma classe versus todas as outras. Depois de implementar um único neurônio, eu vou criar vários para fazer predições com datasets com mais de uma classe.

#### Função para adicionar 1 ao final dos dados (BIAS)

```python
def adicionarDimensaoBias(x):
    
    x = np.append(x, np.ones((len(x), 1)), axis=1)
    
    return x
```

```python
print(xTrain[0])
print(xTrain[0].shape)
data = adicionarDimensaoBias(xTrain)
print(xTrain[0])
print(xTrain[0].shape)
```

#### Função para iniciar os pesos aleatórios de um neurônio

Será necessário enviar o tamanho do vetor JÁ CONTANDO COM O BIAS!

```python
def iniciarPesos(tamanho):
    pesos = []
    
    for i in range(tamanho):
        pesos.append(randint(-1000, 1000) / 1000)
        
    return np.array(pesos)
```

```python
tamanho = len(xTrain[0]) # JA TEM O BIAS
pesos   = iniciarPesos(tamanho)
print(pesos)
print(pesos.shape)
```

#### Função para calcular a entrada líquida de um neurônio dado o vetor de pesos e o dado de entrada

```python
def calcularEntradaLiquida(pesos, dado):
    return np.matmul(pesos.T, dado)
```

```python
entradaLiquida = calcularEntradaLiquida(pesos, xTrain[0])
print(entradaLiquida)
```

#### Função de ativação do neurônio

```python
def calcularFuncaoAtivacao(entradaLiquida):
    if entradaLiquida >= 0:
        return 1
    else:
        return 0
```

```python
calcularFuncaoAtivacao(entradaLiquida)
```

#### Função para predizer um único dado

Como o neurônio é binário, essa função só vai dizer se o dado pertence à classe 1 ou à classe 0. Ela só vai unir as funções de entrada líquida e de ativação.

```python
def predizerUnicoDado(pesos, dado):
    return calcularFuncaoAtivacao(calcularEntradaLiquida(pesos, dado))
```

```python
predizerUnicoDado(pesos, xTrain[0])
```

#### Função para calcular a acurácia do neurônio

```python
def calcularAcuracia(x, yReal, pesos):
        
        qtdAcertos = 0
        
        for dadoAtual, classeReal in zip(x, yReal):
            classePredita = predizerUnicoDado(pesos, dadoAtual)
            if classePredita == classeReal:
                qtdAcertos += 1

        return qtdAcertos/len(x)
```

```python
calcularAcuracia(xTrain, yTrain, pesos)
```

#### Função para realizar uma época e atualizando os pesos do neurônio

```python
def atualizarPesosUnicaEpoca(xTreinamento, yTreinamento, pesos, taxaAprendizagem):
    
    # PASSANDO POR CADA DADO
    for dadoAtual, classeReal in zip(xTreinamento, yTreinamento):

        # FAZENDO A PREDICAO DO DADO ATUAL
        classePredita = predizerUnicoDado(pesos, dadoAtual)
        
        # CALCULANDO O ERRO
        erro = classeReal - classePredita

        # ATUALIZANDO O VETOR DE PESOS
        pesos = pesos + taxaAprendizagem * erro * dadoAtual
        
    return pesos
```

```python
print(pesos)
print(pesos.shape)
pesos = atualizarPesosUnicaEpoca(xTrain, yTrain, pesos, 0.001)
print(pesos)
print(pesos.shape)
```

#### Função para ajustar os pesos do neurônio época após época

É a função que vai ficar responsável por iterar cada época. Ela também dará o critério de parada.

```python
def iterarEpocasTreinamentoNeuronio(xTreinamento, yTreinamento, pesos, taxaAprendizagem=1e-3, qtdMaxEpocas=1000, armazenarEvolucao=True):
    
    # ARRAYS PARA ARMAZENAR A EVOLUCAO DE CADA EPOCA (VOU PRECISAR PLOTAR ISSO DEPOIS).
    arrayAcuracias = [] # ARRAY DE ACURACIAS AO LONGO DAS EPOCAS
    matrizEvolucaoPesos = [] # CADA LINHA UMA EPOCA, CADA COLUNA UM PESO
    
    # NO FINAL, EU VOU RETORNAR OS PESOS QUE TIVERAM A MELHOR ACURACIA
    melhorAcuracia = 0
    melhoresPesos  = np.zeros(len(pesos))
    
    # TREINANDO POR VARIAS EPOCAS. A EPOCA 0 SAO OS PESOS ALEATORIOS
    epocas = -1
    while epocas <= qtdMaxEpocas:
        epocas += 1
        
        # COMECO VERIFICANDO A ACURACIA DESSE NEURONIO
        acuraciaAtual = calcularAcuracia(xTreinamento, yTreinamento, pesos)
        
        # ARMAZENO A EVOLUCAO DA EPOCA
        if armazenarEvolucao == True:
            arrayAcuracias.append(acuraciaAtual)
            matrizEvolucaoPesos.append(pesos)
            
        # VERIFICO SE FOI A MELHOR ITERACAO ATE AGORA
        if acuraciaAtual > melhorAcuracia:
            melhorAcuracia = acuraciaAtual
            melhoresPesos  = pesos
    
        # SE TIVER CONVERGIDO EU PARO
        if acuraciaAtual == 1:
            break
            
        # SE NAO TIVER CONVERGIDO EU CONTINUO O ALGORITMO E AJUSTO OS PESOS DESSA EPOCA
        pesos = atualizarPesosUnicaEpoca(xTreinamento, yTreinamento, pesos, taxaAprendizagem)
        
    # TRANSFORMANDO OS ARRAYS EM NUMPY
    # VOU TRANSPOR A MATRIZ DE EVOLUCAO DOS PESOS, ASSIM, CADA LINHA E UM PESO E CADA COLUNA UMA EPOCA
    epocas              = np.array(epocas)
    arrayAcuracias      = np.array(arrayAcuracias)
    matrizEvolucaoPesos = np.array(matrizEvolucaoPesos).T 
        
    return melhoresPesos, epocas, arrayAcuracias, matrizEvolucaoPesos
```

```python
pesos, epocas, evolucaoAcuracias, evolucaoPesos = iterarEpocasTreinamentoNeuronio(xTrain, yTrain, pesos)
```

```python
print("Acurácia máxima de treinamento:", calcularAcuracia(xTrain, yTrain, pesos))
print("Acurácia de teste:", calcularAcuracia(xTest, yTest, pesos))
plt.plot(evolucaoAcuracias)
plt.grid(alpha=0.5)
plt.title("Evolução da acurácia")
plt.xlabel("Época")
plt.ylabel("Acurácia")
```

#### Função para predizer todos os dados do conjutno de teste

```python
def predizer(xTeste, pesos):
        
        # ADICIONANDO UMA DIMENSAO PARA O BIAS
        xTeste = adicionarDimensaoBias(xTeste)
        
        print(xTeste.shape)
        
        # PREDIZENDO CADA DADO
        yPred = []        
        for dadoAtual in xTeste:
            print(dadoAtual.shape)
            print(pesos.shape)
            yPred.append(predizerUnicoDado(pesos, dadoAtual))
            
        return yPred
```

```python
yPred = predizer(xTest, pesos)
accuracy_score(target, yPred)
```

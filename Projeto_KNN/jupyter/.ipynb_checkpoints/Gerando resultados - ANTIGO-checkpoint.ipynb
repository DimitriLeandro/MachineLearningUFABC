```python
#DEFININDO ALGUNS PARÂMETROS PARA PLOTAR GRÁFICOS
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
params = {'figure.figsize': [10, 5], 
          'axes.labelsize': 10,
          'axes.titlesize':14, 
          'font.size': 10,
          'legend.fontsize': 10, 
          'xtick.labelsize': 8, 
          'ytick.labelsize': 8
         }
plt.rcParams.update(params)
```

```python
def plotar(nomeDataset, matrizAcuraciasDimi, matrizAcuraciasSKLearn, matrizTemposDimi, matrizTemposSKLearn):
    # RECUPERANDO A MEDIA E O DESVIO PADRAO PARA CADA VALOR DE K
    mediasAcuraciasDimi      = [np.mean(acuraciasKAtual) for acuraciasKAtual in matrizAcuraciasDimi]
    desvPadsAcuraciasDimi    = [np.std(acuraciasKAtual) for acuraciasKAtual in matrizAcuraciasDimi]
    mediasAcuraciasSKLearn   = [np.mean(acuraciasKAtual) for acuraciasKAtual in matrizAcuraciasSKLearn]
    desvPadsAcuraciasSKLearn = [np.std(acuraciasKAtual) for acuraciasKAtual in matrizAcuraciasSKLearn]

    mediasTemposDimi      = [np.mean(temposKAtual) for temposKAtual in matrizTemposDimi]
    desvPadsTemposDimi    = [np.std(temposKAtual) for temposKAtual in matrizTemposDimi]
    mediasTemposSKLearn   = [np.mean(temposKAtual) for temposKAtual in matrizTemposSKLearn]
    desvPadsTemposSKLearn = [np.std(temposKAtual) for temposKAtual in matrizTemposSKLearn]


    # PLOTANDO
    fig, ax1 = plt.subplots()

    ax1.errorbar(arrayValoresK, mediasAcuraciasDimi, yerr=desvPadsAcuraciasDimi, fmt=".--", linewidth=1, elinewidth=1, capsize=2, label="Acurácia KNN Próprio", color="#ffff00")
    ax1.errorbar(arrayValoresK, mediasAcuraciasSKLearn, yerr=desvPadsAcuraciasSKLearn, fmt=".--", linewidth=1, elinewidth=1, capsize=2, label="Acurácia Scikit Learn", color="#0000ff")
    ax1.set_xlabel("Valor de K")
    ax1.set_ylabel("Acurácia")

    ax2 = ax1.twinx()

    ax2.errorbar(arrayValoresK, mediasTemposDimi, yerr=desvPadsTemposDimi, fmt=".--", linewidth=1, elinewidth=1, capsize=2, label="Tempo KNN Próprio", color="#00ff00")
    ax2.errorbar(arrayValoresK, mediasTemposSKLearn, yerr=desvPadsTemposSKLearn, fmt=".--", linewidth=1, elinewidth=1, capsize=2, label="Tempo Scikit Learn", color="#ff0000")
    ax2.set_ylabel("Tempo de Processamento (s)")

    plt.title("Resultados para o " + nomeDataset)
    h1, l1 = ax1.get_legend_handles_labels()
    h2, l2 = ax2.get_legend_handles_labels()
    ax1.legend(h1+h2, l1+l2, loc='upper left')
    plt.xticks(arrayValoresK)
    plt.grid(alpha=0.25)
    plt.tight_layout()
    
    caminhoSalvarPDF = "../resultados/resultado_" + nomeDataset.replace(" ", "") + ".pfd"
    plt.savefig(caminhoSalvarPDF, format='pdf', bbox_inches = 'tight')
    
    plt.show()
```

### Dataset 1

```python
%%time

# PEGO OS DADOS DO DATASET COMO ARRAYS E NAO PANDAS
x = dataset1[["feature1", "feature2"]].values
y = dataset1["classe"].values

# MATRIZ DE ACURACIAS E TEMPOS -> CADA LINHA E UM VALOR DE K E CADA COLUNA E UMA INTERACAO DO HOLDOUT
matrizAcuraciasDataset1Dimi     = []
matrizAcuraciasDataset1SKLearn  = []
matrizTemposDataset1Dimi        = []
matrizTemposDataset1SKLearn     = []

# DEFININDO O RANGE DE K
arrayValoresK = np.arange(1, 10, 2)

# PARA CADA VALOR DE K
for kAtual in arrayValoresK:
    
    linhaAcuraciasKAtualDimi    = []
    linhaAcuraciasKAtualSKLearn = []
    linhaTemposKAtualDimi       = []
    linhaTemposKAtualSKLearn    = []
    
    # FACO UM BOOTSTRAP COM 5 HOLDOUTS
    for crossValidation in range(10):
        
        # SEPARANDO ENTRE DADOS DE TREINAMENTO E TESTE
        xTrain, xTest, yTrain, yTest = tts(x, y, test_size=0.25, shuffle=True)
        
        # UTILIZANDO A CLASSE CRIADA PARA O KNN
        inicioDimi = time()
        objKNNDimi = KNN(kAtual)
        objKNNDimi.treinar(xTrain, yTrain)
        yPredDimi  = objKNNDimi.predizer(xTest)
        fimDimi = time()
        
        # UTILIZANDO O SKLEARN
        inicioSKLearn = time()
        objKNNSKLearn = KNeighborsClassifier(n_neighbors=kAtual).fit(xTrain, yTrain)
        yPredSKLearn  = objKNNSKLearn.predict(xTest)
        fimSKLearn    = time()
        
        # COLOCANDO O RESULTADO NA LINHA DO K ATUAL NA MATRIZ DE ACURACIAS E TEMPOS
        linhaAcuraciasKAtualDimi.append(accuracy_score(yTest, yPredDimi))
        linhaAcuraciasKAtualSKLearn.append(accuracy_score(yTest, yPredSKLearn))
        linhaTemposKAtualDimi.append(fimDimi - inicioDimi)
        linhaTemposKAtualSKLearn.append(fimSKLearn - inicioSKLearn)
        
    # AGORA QUE JA TENHO TODOS OS RESULTADOS DO HOLDOUT PARA O K ATUAL, COLOCO A NOVA LINHA NA MATRIZ
    matrizAcuraciasDataset1Dimi.append(linhaAcuraciasKAtualDimi)
    matrizAcuraciasDataset1SKLearn.append(linhaAcuraciasKAtualSKLearn)
    matrizTemposDataset1Dimi.append(linhaTemposKAtualDimi)
    matrizTemposDataset1SKLearn.append(linhaTemposKAtualSKLearn)
```

```python
plotar("Dataset 1", matrizAcuraciasDataset1Dimi, matrizAcuraciasDataset1SKLearn, matrizTemposDataset1Dimi, matrizTemposDataset1SKLearn)
```

### Cluster in Cluster

```python
%%time

# PEGO OS DADOS DO DATASET COMO ARRAYS E NAO PANDAS
x = dataset1b[["feature1", "feature2"]].values
y = dataset1b["classe"].values

# MATRIZ DE ACURACIAS E TEMPOS -> CADA LINHA E UM VALOR DE K E CADA COLUNA E UMA INTERACAO DO HOLDOUT
matrizAcuraciasDataset1bDimi     = []
matrizAcuraciasDataset1bSKLearn  = []
matrizTemposDataset1bDimi        = []
matrizTemposDataset1bSKLearn     = []

# DEFININDO O RANGE DE K
arrayValoresK = np.arange(1, 20, 2)

# PARA CADA VALOR DE K
for kAtual in arrayValoresK:
    
    linhaAcuraciasKAtualDimi    = []
    linhaAcuraciasKAtualSKLearn = []
    linhaTemposKAtualDimi       = []
    linhaTemposKAtualSKLearn    = []
    
    # FACO UM BOOTSTRAP COM 5 HOLDOUTS
    for crossValidation in range(10):
        
        # SEPARANDO ENTRE DADOS DE TREINAMENTO E TESTE
        xTrain, xTest, yTrain, yTest = tts(x, y, test_size=0.25, shuffle=True)
        
        # UTILIZANDO A CLASSE CRIADA PARA O KNN
        inicioDimi = time()
        objKNNDimi = KNN(kAtual)
        objKNNDimi.treinar(xTrain, yTrain)
        yPredDimi  = objKNNDimi.predizer(xTest)
        fimDimi = time()
        
        # UTILIZANDO O SKLEARN
        inicioSKLearn = time()
        objKNNSKLearn = KNeighborsClassifier(n_neighbors=kAtual).fit(xTrain, yTrain)
        yPredSKLearn  = objKNNSKLearn.predict(xTest)
        fimSKLearn    = time()
        
        # COLOCANDO O RESULTADO NA LINHA DO K ATUAL NA MATRIZ DE ACURACIAS E TEMPOS
        linhaAcuraciasKAtualDimi.append(accuracy_score(yTest, yPredDimi))
        linhaAcuraciasKAtualSKLearn.append(accuracy_score(yTest, yPredSKLearn))
        linhaTemposKAtualDimi.append(fimDimi - inicioDimi)
        linhaTemposKAtualSKLearn.append(fimSKLearn - inicioSKLearn)
        
    # AGORA QUE JA TENHO TODOS OS RESULTADOS DO HOLDOUT PARA O K ATUAL, COLOCO A NOVA LINHA NA MATRIZ
    matrizAcuraciasDataset1bDimi.append(linhaAcuraciasKAtualDimi)
    matrizAcuraciasDataset1bSKLearn.append(linhaAcuraciasKAtualSKLearn)
    matrizTemposDataset1bDimi.append(linhaTemposKAtualDimi)
    matrizTemposDataset1bSKLearn.append(linhaTemposKAtualSKLearn)
```

```python
plotar("Cluster in Cluster", matrizAcuraciasDataset1bDimi, matrizAcuraciasDataset1bSKLearn, matrizTemposDataset1bDimi, matrizTemposDataset1bSKLearn)
```
